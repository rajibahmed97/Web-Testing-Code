# Plotting with keyword string

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c = 'c', s = 'd', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
# plt.tight_layout()
plt.show()
```

```python
data = {
    'a':np.arange(50),
    'b':np.random.randint(0,50,50),
    'c':np.random.randn(50) * np.random.randn(50)
}
data['d'] = np.random.randint(0,50,50)
data['e'] = data['a'] + 10 ** np.random.randn(50)
data['d'] = data['b'] + np.random.randint(0,50,50)

plt.scatter('a', 'b', 'c', data=data)
plt.xlabel('x label')
plt.ylabel('y label')
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c = 'c', s = 'd', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.tight_layout()
plt.show()
```
